```python
import pandas as pd
#import csv
import streamlit as st
import numpy as np
import yfinance as yf
import datetime
import matplotlib.pyplot as plt
#import os
```

```python
nifty = pd.read_csv("https://www1.nseindia.com/content/indices/ind_nifty500list.csv")
#@https://www1.nseindia.com/content/indices/ind_nifty50list.csv")
nifty
```

```python
nifty["Symbol"] = nifty["Symbol"]+".NS"
```

```python
index = "^CRSLDX"
indexdata = yf.download(index, period="10y")
```

```python
data = yf.download(nifty["Symbol"].tolist(), period="10y")
```

```python
data
```

```python
##Mean of % above 200 and 50 DMA
dma200=pd.DataFrame()
dma50=pd.DataFrame()
for i in range(len(nifty["Symbol"].tolist())):
    stock = nifty["Symbol"].iloc[i]
    data1 = data["Adj Close"][stock].reset_index()#yf.download(stock, period="5y")
    data1["200dma"]=data1[stock].rolling(window=200).mean()
    data1["50dma"]=data1[stock].rolling(window=50).mean()
    data1["Check200"]=data1[stock]/data1["200dma"]
    data1["Check50"]=data1[stock]/data1["50dma"]
    dma200[stock]=data1["Check200"]
    dma50[stock]=data1["Check50"]
```

```python
data1
```

```python
dma200["Date"] = data1["Date"]
dma50["Date"] = data1["Date"]
```

```python
df1=pd.DataFrame()
df1["200"]=dma200.median(axis=1)
df1["50"]=dma50.median(axis=1)
#df1.reset_index(inplace=True)
df1["Date"] = data1["Date"]
df1
```

```python
df1=pd.DataFrame()
df1["200"]=dma200.sum(axis=1)/500
df1["50"]=dma50.sum(axis=1)/500
#df1.reset_index(inplace=True)
df1["Date"] = data1["Date"]
df1
```

```python
ref_date = "2013-01-01"
df2 = df1[df1["Date"]>ref_date]
```

```python
from matplotlib.dates import DateFormatter
fig, ax = plt.subplots(figsize=(20,13))
plt.rcParams.update({'font.size': 18})
ax.set_title("How much over 200 and 50 DMA are stocks trading - median")
ax.plot(df2["Date"], df2["200"], label="200-dma")
ax.plot(df2["Date"], df2["50"], label="50-dma")
#ax.plot(df1["50"], label="50-dma")
ax2 = ax.twinx()
ax2.plot(indexdata[indexdata.index>ref_date]["Adj Close"], label="Index", color="black")
ax.legend(loc='upper left')
ax2.legend(loc = 'upper right')
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.1%}'.format(x) for x in vals])
ax.xaxis.set_major_formatter(DateFormatter("%b-%y"))
plt.show()
```

```python
dma200=pd.DataFrame()
dma50=pd.DataFrame()
for i in range(len(nifty["Symbol"].tolist())):
    stock = nifty["Symbol"].iloc[i]
    data1 = data["Adj Close"][stock].reset_index()#yf.download(stock, period="5y")
    data1["200dma"]=data1[stock].rolling(window=200).mean()
    data1["50dma"]=data1[stock].rolling(window=50).mean()
    data1["Check200"]=np.where(data1[stock]>data1["200dma"],1,0)
    data1["Check50"]=np.where(data1[stock]>data1["50dma"],1,0)
    dma200[stock]=data1["Check200"]
    dma50[stock]=data1["Check50"]
```

```python
n500 = nifty.copy()
#Get latest prices
l = data.iloc[-1]["Adj Close"]
#YTD returns
ytd_prices = data.iloc[data.index.get_loc(datetime.datetime(data.index[-1].year,1,1),method='nearest')]["Adj Close"]
df = (l/ytd_prices - 1).reset_index()
df.columns = ["Symbol", "YTD ret"]
n500 = n500.merge(df, on="Symbol")
#1Y returns
y1_prices = data.iloc[data.index.get_loc(data.index[-1] - np.timedelta64(1, "Y"),method='nearest')]["Adj Close"]
df = (l/y1_prices - 1).reset_index()
df.columns = ["Symbol", "1Y ret"]
n500 = n500.merge(df, on="Symbol")
#5Y returns
y5_prices = data.iloc[data.index.get_loc(data.index[-1] - np.timedelta64(5, "Y"),method='nearest')]["Adj Close"]
df = (l/y5_prices - 1).reset_index()
df.columns = ["Symbol", "5Y ret"]
n500 = n500.merge(df, on="Symbol")
(n500.groupby("Industry").median()*100).round(2)
```
